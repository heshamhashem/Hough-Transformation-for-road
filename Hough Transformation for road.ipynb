### In this project we will build Hough Transformation for road 


```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('test_image.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
blur = cv2.GaussianBlur(img,(7,7),20)           #Gaussian Bluring filter
gray = cv2.cvtColor(blur,cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(gray,50,120)                  #canny edgedetection
plt.figure(figsize=(16, 12))
plt.subplot(1,2,1),plt.imshow(img)
plt.grid()
plt.title('Detected')
plt.subplot(1,2,2),plt.imshow(edges,cmap = 'gray')
plt.title('Edges')
```

```python
img = cv2.imread('test_image.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
blur = cv2.GaussianBlur(img,(7,7),20) 
gray = cv2.cvtColor(blur,cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(gray,50,120)

mask=np.zeros_like(edges)

contours = np.array([[(200,700),(500,300),(625,300),(990,700)]])

poly=cv2.fillPoly(mask, pts = [contours], color =(255,255,255))

res= cv2.bitwise_and(poly,poly,mask= edges)



lines = cv2.HoughLinesP(res,1,np.pi/180,100,minLineLength=100,maxLineGap=10)

for line in lines:
    x1,y1,x2,y2 = line[0] 
    cv2.line(img,(x1,y1),(x2,y2),(255,0,0),10)
    
    
plt.figure(figsize=(16, 12))
plt.subplot(1,2,1),plt.imshow(img)


plt.subplot(1,2,2),plt.imshow(res,cmap = 'gray')
```
